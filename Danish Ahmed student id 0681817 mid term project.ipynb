## Introduction to ECG:

. ECG is known as electro cardio gram or also known as electro cardio graph (EKG) is a representation of the electrical events of the cardiac cycles.
. Each event has a distinctive wave form.

ECG Signal : ECG signal can be  read  as QPRST waves 
•	P wave: P wave give us electrical activity of atrial with the help of SA node and we use electrode to calculate electrical activity of heart.
•	QRS waves: QRS waves give us information about ventricular depolarization or contraction activity of heart.
•	T wave:  t waves gives us a result of ventricular repolarization and relaxation activity of heart.


##  b) Working principle of ECG  :
when we perform ECG process to body we placed 12 electrodes on human 6 on the chest of  of the body called chest lead .

Bipolar limb leads (frontal plane):

Lead I: RA (-) to LA (+) (Right Left, or lateral)
Lead II: RA (-) to LL (+) (Superior Inferior)
Lead III: LA (-) to LL (+) (Superior Inferior)
Augmented unipolar limb leads (frontal plane):

Lead aVR: RA (+) to [LA & LL] (-) (Rightward)
Lead aVL: LA (+) to [RA & LL] (-) (Leftward)
Lead aVF: LL (+) to [RA & LA] (-) (Inferior)
Unipolar (+) chest leads (horizontal plane):

Leads V1, V2, V3: (Posterior Anterior)
Leads V4, V5, V6:(Right Left, or lateral)
 
 so with the help of these leads we calculated the electrical activity of heart and after that we got a signal in the form of qprst through these waves we can analyze the result according to the nature of waves

## question 14 
In this question designing a low order buttterworth filter . from this type of butter worth filter i am searching calculator of butter worth designing and so in refrence theres a link of calculating fc1, fc2, fc3 ,fc4 at 100 Hz you gives us but for calculate we convert 100 Hz to 0.1 KHz after that we obtained values of 4th order butter worth low-pass filter 

```python
from scipy import signal

#Getting it ready to filter 100Hz
fs=1.3e-3;
fc1=1.72e3;
fc2=1.47e3;
fc3=4.15e3;
fc4=0.609e3;

b, a = signal.butter(4, [fc1/fs,fc2/fs,fc3/fs,fc4/fs], 'bandstop', analog=True)
w, h = signal.freqs(b, a)
wf=w*fs;
plt.plot(wf, 20 * np.log10(abs(h)))
plt.xscale('log')
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='green') # cutoff frequency
plt.show()
```

# Question 9 :
In question 9 dft signal and plot full spectrum .

```python
#Adding the noise
ns = 500*np.random.normal(0,1,15000)
t=np.arange(0,12e-6,0.0008e-6) #time vector
AllSignalN = 50*np.sin(2*np.pi*1e6*t)+100*np.sin(2*np.pi*2e6*t)+150*np.sin(2*np.pi*3e6*t)+40*np.sin(2*np.pi*5e6*t) #new composed signal
AllSignalNs=AllSignalN+ns
#Plotting the new signal
plt.subplot(221)
plt.tight_layout()
plt.plot(t,AllSignalNs)
plt.title('New Composed Signal with Noise')
plt.xlabel('Time (us)')
plt.ylabel('Amplitude')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
L=len(AllSignalNs)
f=np.arange(0,1.3e-9/2,1.3e-9/15000)
Y=np.fft.fft(AllSignalNs)
Pyy=abs(Y)/L
print(len(Pyy))
plt.subplot(222)
plt.plot(f, Pyy[:7500])
plt.title('DFT of new signal with noise')
plt.ylabel('Amplitude')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
```

## question 16 :
Obtain the DFT of the filtered signal and plot the full spectrum after filtering

```python
# filter  signal plot the full spectrum
FiltSignal=signal.lfilter(b,a,AllSignalNs)
#Plotting the new signal
plt.subplot(221)
plt.tight_layout()
plt.plot(t,FiltSignal)
plt.title('New Composed Signal with Noise and Filtered at 5MHz')
plt.xlabel('Time (us)')
plt.ylabel('Amplitude')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
L=len(FiltSignal)
f=np.arange(0,1.25e9/2,1.25e9/15000)
Y=np.fft.fft(FiltSignal)
Pyy=abs(Y)/L
print(len(Pyy))
plt.subplot(222)
plt.plot(f, Pyy[:7500])
plt.title('DFT of new signal with noise')
plt.ylabel('Amplitude')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
#Zooming into just less than 15MHz
inx=np.where(f>1.3e-3)
inx=inx[0]
plt.subplot(223)
plt.plot(f[:inx[0]], Pyy[:inx[0]])
plt.title('DFT of new signal zoomed with noise')
plt.ylabel('Amplitude')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
```
